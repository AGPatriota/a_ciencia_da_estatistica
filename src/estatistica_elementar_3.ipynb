### Gráficos de barras e pizza

```r
x    = c(682, 1434, 32, 237, 1702, 4172, 1257, 590, 23)
col0=c("#23b796","#27CCA7","#22B292","#1D997D","#1b8e74","#1FA585","#136653","#0E4C3E","#093329") 
Es   = c("AC", "AM", "AP", "MA", "MT", "PA", "RO", "RR", "TO")
options(OutDec= ",") 
d1    = barplot(x,names.arg=Es,main="DESMATAMENTO EM KM2",col="#0E4C3E",ylim=c(0,4500))
text(d1[,1],x+200,x)
d2    = barplot(x/sum(x)*100,names.arg=Es,main="DESMATAMENTO EM %",col="#0E4C3E",ylim=c(0,50))
text(d2[,1],x/sum(x)*100+2,paste(round(x/sum(x)*100,2),"%",sep=""))
pie(x/sum(x)*100,labels=Es,main="DESMATAMENTO EM %", col= col0)
```

## Gráficos de dispersão

```r
x = c(27772, 19014, 14286, 11651, 12911, 7464, 7000, 6418, 4571, 5891, 5012, 6207, 7893, 6947, 7536, 10129)
plot(x~c(2004:2019), main="DESMATAMENTO EM KM2", pch=19, xlab="", ylab="", col="#0E4C3E")
plot(x~c(2004:2019), main="DESMATAMENTO EM KM2", xlab="", ylab="", 
col="#0E4C3E", type="l", lwd=4)     
points(x~c(2004:2019), col="#0E4C3E", pch=19)                          
plot(x~c(2004:2019), main="DESMATAMENTO EM KM2", xlab="", ylab="", 
col="#0E4C3E", type="h", lwd=4)
points(x~c(2004:2019), col="#0E4C3E", pch=19)                          
```
